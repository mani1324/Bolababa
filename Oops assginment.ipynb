```python
##16. How do you prevent a class from having multiple instances by using constructors in Python? Provide an example.

'''
To prevent a class from having multiple instances, you can utilize a design pattern called the Singleton pattern. This pattern restricts
a class to have only one instance by controlling the instantiation process and providing a way to access that single instance. In Python, 
this can be achieved by overriding the __new__ method of the class to control object creation and store the single instance within the class.
'''
##Here's an example of implementing the Singleton pattern:

class Singleton:
    _instance = None

    def __new__(cls):
        if not cls._instance:
            cls._instance = super(Singleton, cls).__new__(cls)
        return cls._instance


instance1 = Singleton()
instance2 = Singleton()

print(instance1 is instance2) 
```

```python
##19. Explain the use of constructor chaining in Python. Provide a practical example.
'''
 Constructor chaining, also known as method resolution order (MRO), refers to the ability of a subclass to use constructors of both its
 own class and its parent class. In Python, when a subclass defines its own constructor (__init__), it needs to explicitly call the constructor 
 of the parent class to ensure proper initialization of inherited attributes. This can be achieved using super().
'''
class Parent:
    def __init__(self, parent_attribute):
        self.parent_attribute = parent_attribute
        print("Parent constructor called")

class Child(Parent):
    def __init__(self, parent_attribute, child_attribute):
        super().__init__(parent_attribute)  
        self.child_attribute = child_attribute
        print("Child constructor called")

child_obj = Child("Parent data", "Child data")
```

```python
##1. What is inheritance in Python? Explain its significance in object-oriented programming.
'''
 Inheritance in Python is a fundamental concept in object-oriented programming (OOP) that allows a class to inherit attributes 
 and methods from another class. It is a mechanism that promotes code reuse and the creation of a hierarchy of classes where a
 new class (the subclass or derived class) can inherit properties and behaviors from an existing class (the superclass or base class). 
 Inheritance forms an "is-a" relationship between classes, where a subclass is a specialized version of its superclass.

 Key points about inheritance in Python and its significance in OOP:

 1. Base Class (Superclass): The class from which attributes and methods are inherited is called the superclass or base class.

 2. Derived Class (Subclass): The class that inherits attributes and methods is called the subclass or derived class.

 3. Inherited Features: Subclasses inherit attributes and methods from the superclass, which they can use, override, or extend as needed.

 4. Code Reusability: Inheritance promotes code reusability by allowing common attributes and behaviors to be defined in the superclass and reused in
    multiple subclasses.

 5. Method Overriding: Subclasses can override (modify) methods inherited from the superclass to provide specialized behavior while still retaining
    the interface.
 
 6. Polymorphism: Inheritance is a key component of polymorphism, allowing objects of different classes to be treated as objects of their common superclass.

 7. Hierarchical Structure: Inheritance enables the creation of a hierarchy of classes, with more specialized classes extending the functionality of
    more general classes.

 8. is-a Relationship: Inheritance models an "is-a" relationship, where a subclass "is a kind of" its superclass.
 
'''
class Animal:
    def __init__(self, species):
        self.species = species

    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"
    
dog = Dog("Canine")
cat = Cat("Feline")

print(f"A {dog.species} says: {dog.speak()}")
print(f"A {cat.species} says: {cat.speak()}")
```

```python
##5. How can you access the methods and attributes of a parent class from a child class in Python? Give an example.

'''
 In Python, you can access the methods and attributes of a parent class from a child class using the super() method or
 by directly referencing the parent class itself.

 Example using super():
 Here's an example demonstrating how to access the parent class's methods and attributes using super():
'''
class Parent:
    def __init__(self):
        self.parent_attribute = "I'm from the parent class"

    def parent_method(self):
        return "Method from parent class"

class Child(Parent):
    def __init__(self):
        super().__init__()

    def access_parent(self):
        return super().parent_method(), self.parent_attribute

child = Child()

method_result, attribute = child.access_parent()
print(method_result)  
print(attribute)      

'''
 Example using direct reference:
 Here's the same functionality, but with direct reference to the parent class's methods and attributes:
 
'''
class Parent:
    def __init__(self):
        self.parent_attribute = "I'm from the parent class"

    def parent_method(self):
        return "Method from parent class"

class Child(Parent):
    def __init__(self):
        Parent.__init__(self)

    def access_parent(self):
        return self.parent_method(), self.parent_attribute

child = Child()

method_result, attribute = child.access_parent()
print(method_result) 
print(attribute)      
```

```python
'''11. Create a Python class called `Shape` with a method `area()` that calculates the area of a shape. Then, create child classes `Circle` and
`Rectangle` that inherit from `Shape` and implement the `area()` method accordingly. Provide an example.'''

import math

class Shape:
    def area(self):
        return 0  

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2  

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height  

circle = Circle(5)
rectangle = Rectangle(4, 6)

print("Area of the circle:", circle.area())       
print("Area of the rectangle:", rectangle.area()) 
```

```python
##12. Explain the use of abstract base classes (ABCs) in Python and how they relate to inheritance. Provide an example using the `abc` module.

'''
Abstract Base Classes (ABCs) in Python, found in the abc (Abstract Base Classes) module, provide a way to define abstract methods that 
must be implemented by their subclasses. They help to define a blueprint or interface for a class without necessarily providing a complete 
implementation, allowing for a more structured approach to inheritance.

Use of Abstract Base Classes (ABCs):
Defining Abstract Methods: ABCs define abstract methods which serve as placeholders, outlining methods that must be implemented in the subclasses.
Forcing Implementation: Subclasses that inherit from an ABC must implement all the abstract methods defined in the ABC, ensuring their presence and functionality.
Example using abc module:
Here's an example demonstrating the use of the abc module to create an abstract class with abstract methods that need to be implemented by the subclasses:

'''
from abc import ABC, abstractmethod

class Shape(ABC):  
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius  

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height  

circle = Circle(5)
rectangle = Rectangle(4, 6)

print("Area of the circle:", circle.area())       
print("Area of the rectangle:", rectangle.area()) 

```

```python
##18. What is the "diamond problem" in multiple inheritance, and how does Python address it?
'''
The "diamond problem" is a challenge encountered in multiple inheritance when a subclass inherits from two classes that have a 
common ancestor. This creates an ambiguity when methods or attributes from the shared ancestor are accessed through the subclass. It is 
named "diamond" because of the inheritance diagram's shape.

Consider the classic diamond inheritance structur:


   A
 /   \
B     C
 \   /
   D

Classes B and C both inherit from class A, and class D inherits from both B and C.
Ambiguity in the Diamond Problem:
Method Conflict: If classes B and C each have a method some_method(), and class D calls some_method(), it's ambiguous which 
method should be inherited since both B and C inherit from A with the same method name.

Attribute Conflict: Similarly, if both B and C define the same attribute, class D faces ambiguity in selecting the attribute to inherit.

Python's Approach to the Diamond Problem:
Python addresses the diamond problem using a mechanism called Method Resolution Order (MRO), specifically the C3 Linearization 
algorithm. Python's MRO defines the order in which methods and attributes are searched for and resolved in multiple inheritance.

The MRO algorithm ensures that the method resolution follows a specific order, primarily the order in which the base classes are defined.
It uses a depth-first, left-to-right search to ensure a consistent and unambiguous method resolution order.

'''
#Example:

class A:
    def some_method(self):
        print("Method from A")

class B(A):
    pass

class C(A):
    def some_method(self):
        print("Method from C")

class D(B, C):
    pass

# Output will be "Method from C" due to Python's MRO
obj = D()
obj.some_method()
```

```python
##15. Describe the concept of operator overloading in Python and how it relates to polymorphism. Provide examples using operators like `+` and `*`.
'''
Operator overloading in Python refers to the ability to define and redefine how operators behave for objects of a custom class.
This concept is closely related to polymorphism, where different objects can respond to the same operator in a way that makes sense
for their specific context.

In Python, operator overloading is achieved by defining special methods in a class. These methods have double underscores (__) as prefixes
and correspond to specific operators. For example, the __add__ method is used to overload the + operator, and the __mul__ method is used to
overload the * operator.

Let's look at examples using the + and * operators:

'''
#Example 1: Overloading the + Operator

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Point):
            # Add corresponding coordinates
            return Point(self.x + other.x, self.y + other.y)
        else:
            raise TypeError("Unsupported operand type. Must be a Point object.")

# Usage
point1 = Point(1, 2)
point2 = Point(3, 4)

result = point1 + point2
print(result.x, result.y)  


#Example 2: Overloading the * Operator

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __mul__(self, scalar):
        # Multiply each coordinate by the scalar
        return Vector(self.x * scalar, self.y * scalar)

# Usage
vector = Vector(2, 3)

result = vector * 3
print(result.x, result.y)  
```

```python
##10. What is the purpose of abstract methods, and how do they enforce abstraction in Python classes?
'''
Abstract methods in Python are methods declared in an abstract base class (ABC) that don't have an implementation in the base class.
Instead, their implementation is left to the concrete subclasses that inherit from the abstract base class. The purpose of abstract 
methods is to define a common interface that must be implemented by all the subclasses, enforcing a certain level of abstraction.

Here's how abstract methods enforce abstraction in Python classes:

Forcing Implementation in Subclasses:
  Abstract methods act as placeholders for functionality that must be implemented by any concrete subclass.
  If a subclass fails to provide an implementation for an abstract method, it itself becomes abstract, and an error occurs if an attempt 
  is made to create an instance of that subclass.

Defining a Common Interface:
  Abstract methods define a common interface that all subclasses must adhere to.
  This ensures that subclasses provide specific functionality, contributing to a consistent and predictable API for the entire class hierarchy.
  
'''
#Here's an example illustrating the use of abstract methods for abstraction:
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius**2

    def perimeter(self):
        return 2 * 3.14 * self.radius

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side**2

    def perimeter(self):
        return 4 * self.side

# Usage
circle = Circle(radius=5)
square = Square(side=4)

print(f"Circle Area: {circle.area()}, Perimeter: {circle.perimeter()}")
print(f"Square Area: {square.area()}, Perimeter: {square.perimeter()}")
```

```python
##15. Explain the concept of abstract data types (ADTs) and their role in achieving abstraction in Python.

'''
Abstract Data Types (ADTs) refer to a high-level description of a set of data and the operations that can be performed on that data.
ADTs provide a way to abstract and encapsulate the details of the data structure's implementation, focusing on what operations can be 
performed rather than how they are implemented. In Python, ADTs are typically implemented using classes.

Here are key points about ADTs and their role in achieving abstraction in Python:

High-Level Description:

ADTs describe the behavior of data structures without specifying their implementation details.
They define a set of operations that can be performed on the data, such as insertion, deletion, and retrieval.
Encapsulation:

ADTs encapsulate the data and operations within a single unit (class in Python).
The internal details of how the operations are carried out are hidden from the user.
Abstraction:

ADTs provide abstraction by allowing users to interact with data at a high level, without concerning themselves with the underlying implementation.
Users only need to know what operations are available and how to use them.
Reusability:

ADTs promote code reusability. Once an ADT is defined, it can be used in different parts of a program or even in different programs without modification.
Modularity:

ADTs contribute to code modularity. They enable the separation of concerns, allowing developers to focus on different aspects of a program independently.

'''
from abc import ABC, abstractmethod

class AbstractStack(ABC):
    @abstractmethod
    def push(self, item):
        pass

    @abstractmethod
    def pop(self):
        pass

    @abstractmethod
    def is_empty(self):
        pass

class Stack(AbstractStack):
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

# Usage
stack = Stack()
stack.push(1)
stack.push(2)
print(stack.pop())  # Output: 2
```

```python
##19. Create a Python class for a library system, implementing abstraction by defining common methods (e.g., `add_book()`, `borrow_book()`) in an abstract base class.

from abc import ABC, abstractmethod

# Abstract base class for LibraryItem
class LibraryItem(ABC):
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True

    @abstractmethod
    def display_info(self):
        pass

    @abstractmethod
    def checkout(self):
        pass

    @abstractmethod
    def return_item(self):
        pass

# Concrete class for Book
class Book(LibraryItem):
    def __init__(self, title, author, genre):
        super().__init__(title, author)
        self.genre = genre

    def display_info(self):
        print(f"Book: {self.title} by {self.author} ({self.genre})")

    def checkout(self):
        if self.is_available:
            print(f"Checking out {self.title}")
            self.is_available = False
        else:
            print(f"{self.title} is not available for checkout")

    def return_item(self):
        print(f"Returning {self.title}")
        self.is_available = True

# Concrete class for DVD
class DVD(LibraryItem):
    def __init__(self, title, director, runtime):
        super().__init__(title, director)
        self.runtime = runtime

    def display_info(self):
        print(f"DVD: {self.title} directed by {self.author} ({self.runtime} minutes)")

    def checkout(self):
        if self.is_available:
            print(f"Checking out {self.title}")
            self.is_available = False
        else:
            print(f"{self.title} is not available for checkout")

    def return_item(self):
        print(f"Returning {self.title}")
        self.is_available = True

# Usage
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", "Fiction")
dvd1 = DVD("Inception", "Christopher Nolan", 148)

book1.display_info()
book1.checkout()
book1.return_item()

dvd1.display_info()
dvd1.checkout()
dvd1.return_item()
```

```python
##20. Describe the concept of method abstraction in Python and how it relates to polymorphism.
'''
Method abstraction in Python refers to the practice of defining methods in an abstract manner, without specifying their implementation details,
in abstract classes or interfaces. This concept is closely related to polymorphism, as it allows different classes to provide their own specific 
implementations for the same abstract methods.

Here's how method abstraction relates to polymorphism:

Abstract Methods:
In Python, abstract methods are declared in abstract classes or interfaces using the @abstractmethod decorator.
Abstract methods do not have an implementation in the abstract class; they only provide a method signature.
Subclasses that inherit from the abstract class must provide concrete implementations for these abstract methods.

Polymorphism:
Polymorphism allows objects of different types to be treated as objects of a common base type.
With method abstraction, polymorphism is achieved by defining abstract methods in a base class and providing concrete implementations in its subclasses.
Objects of the subclasses can be used interchangeably based on the common method signatures, promoting code flexibility and reusability.

Method Overriding:
Subclasses that inherit abstract methods are required to override them with their own implementations.
This process is a form of method overriding, where a subclass provides a specific implementation for a method defined in its superclass or abstract base class.

Common Interface:
Abstract methods define a common interface that participating classes must adhere to.
Classes that implement the same abstract methods can be used uniformly, allowing for a consistent and predictable way to interact with objects of different types.


'''
##Here's a simple example to illustrate method abstraction and polymorphism:

from abc import ABC, abstractmethod

# Abstract class with abstract method
class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass

# Concrete classes implementing the abstract method
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

# Polymorphic usage
shapes = [Circle(5), Rectangle(4, 6)]

for shape in shapes:
    print(f"Area: {shape.calculate_area()}")
```

```python
##1. Explain the concept of composition in Python and how it is used to build complex objects from simpler ones.

'''
Composition in Python is a design principle that involves creating complex objects by combining or composing simpler objects. It is a
way to build relationships between classes and allows for the creation of more modular and maintainable code. In composition, one class
contains an instance of another class, and the composed class leverages the functionality of the contained class. This promotes code reuse,
flexibility, and easier maintenance.

Key points about composition in Python:

Object Composition:
Instead of relying on inheritance, composition focuses on creating relationships between classes by including instances of other classes as attributes.

Has-A Relationship:
Composition establishes a "has-a" relationship between classes. For example, if class A has an instance of class B as an attribute, we say that A has a B.

Modularity:
Composition promotes modularity, as changes in one class do not necessarily impact other classes. Each class remains independent, making it easier to 
understand, modify, and extend the code.

Code Reusability:
By composing classes, you can reuse existing code more easily. If a class provides a specific functionality, you can reuse that class by including
an instance of it in another class.

Flexibility:
Composition allows for more flexible and dynamic relationships between objects. You can change the behavior of a class by modifying the composition
without altering the class itself.

Example:

Let's consider a simple example where a Car class is composed of instances of other classes like Engine, Wheel, and FuelTank. Each of these components can
be developed and tested independently.

'''
class Engine:
    def start(self):
        print("Engine starting")

class Wheel:
    def rotate(self):
        print("Wheel rotating")

class FuelTank:
    def refuel(self):
        print("Fuel tank refueling")

class Car:
    def __init__(self):
        self.engine = Engine()
        self.wheels = [Wheel() for _ in range(4)]
        self.fuel_tank = FuelTank()

    def drive(self):
        self.engine.start()
        for wheel in self.wheels:
            wheel.rotate()
        print("Car is moving")

    def refuel(self):
        self.fuel_tank.refuel()
        print("Car refueled")

# Example usage
my_car = Car()
my_car.drive()
my_car.refuel()
```

```python
##7. Explain the concept of "has-a" relationships in composition and how it helps design software systems.

'''
The concept of "has-a" relationships in composition refers to the relationship between classes where one class contains an object of 
another class as a member. This relationship is established by including an instance of one class within another class. The class that 
contains the instance is said to "have" the other class, hence the term "has-a" relationship.

Here are key points about "has-a" relationships in composition and how they benefit software design:

(1). Modularity and Reusability: Composition allows for the creation of modular and reusable components. By breaking down a complex system into smaller,
     independent classes, you can reuse these classes in various contexts.

(2). Flexibility and Maintainability: "Has-a" relationships enhance flexibility in software design. Changes to one class don't necessarily affect 
     other classes as long as the interfaces remain consistent. This makes the system more maintainable and adaptable to changes.

(3). Code Organization: Composition helps in organizing code logically. Each class is responsible for a specific aspect of functionality,
     making the codebase more readable and understandable.

(4). Encapsulation: Composition supports encapsulation by allowing you to hide the internal details of a class. The containing class exposes
     only what is necessary for the outside world, enhancing information hiding.

(5). Loose Coupling: Classes in a "has-a" relationship are loosely coupled, meaning they are not heavily dependent on each other. This reduces
     the impact of changes in one class on others, leading to a more modular and robust system.

(6). Hierarchy Design: By using composition, you can design class hierarchies that are more flexible and extensible. Each class can focus on its 
     specific functionality, and relationships between classes can be easily adjusted.
     
'''
##Here's a simple example to illustrate a "has-a" relationship:

class Engine:
    def start(self):
        return "Engine started"

class Car:
    def __init__(self):
        self.engine = Engine()  # Car has an Engine

    def start(self):
        return f"Car started. {self.engine.start()}"

# Example usage
my_car = Car()
print(my_car.start())  # Output: Car started. Engine started
```

```python
##8. Create a Python class for a computer system, using composition to represent components like CPU, RAM, and storage devices.
class CPU:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def process(self):
        return "Processing data"

class RAM:
    def __init__(self, capacity):
        self.capacity = capacity

    def read_data(self):
        return "Reading data from RAM"

class Storage:
    def __init__(self, capacity):
        self.capacity = capacity

    def store_data(self):
        return "Storing data to storage"

class Computer:
    def __init__(self, cpu, ram, storage):
        self.cpu = cpu  # Computer has a CPU
        self.ram = ram  # Computer has RAM
        self.storage = storage  # Computer has storage

    def execute_program(self):
        processing_result = self.cpu.process()
        reading_result = self.ram.read_data()
        storing_result = self.storage.store_data()
        return f"{processing_result}\n{reading_result}\n{storing_result}"

# Example usage
my_cpu = CPU(brand="Intel", speed="3.0 GHz")
my_ram = RAM(capacity="8 GB")
my_storage = Storage(capacity="1 TB")

my_computer = Computer(cpu=my_cpu, ram=my_ram, storage=my_storage)
result = my_computer.execute_program()

print(result)
```

```python
##9. Describe the concept of "delegation" in composition and how it simplifies the design of complex systems.

'''
Delegation in Composition:

In the context of object-oriented programming and composition, delegation is a design principle where one object hands over a responsibility
to another object. It allows an object to use the functionality of another object, forwarding requests or calls to that object rather than
implementing the functionality itself. Delegation is a way to achieve code reuse and maintainability by distributing responsibilities among objects.

Key Concepts:

Responsibility Transfer: Delegation involves transferring the responsibility for a certain behavior or functionality from one object to another.

Code Reusability: By delegating tasks to other objects, you can reuse existing code without duplicating it in multiple places. 
This promotes a more modular and maintainable codebase.

Encapsulation: Delegation supports encapsulation by allowing an object to hide its internal implementation details while providing a 
clean interface for other objects.


'''
##Example of Delegation:

class PrintingService:
    def print_document(self, document):
        print(f"Printing: {document}")

class Printer:
    def __init__(self, printing_service):
        self.printing_service = printing_service

    def print_document(self, document):
        self.printing_service.print_document(document)

# Example Usage
printing_service = PrintingService()
printer = Printer(printing_service)

printer.print_document("Sample Document")
```

```python
##10. Create a Python class for a car, using composition to represent components like the engine, wheels, and transmission.

class Engine:
    def start(self):
        return "Engine started"

    def stop(self):
        return "Engine stopped"

class Wheels:
    def rotate(self):
        return "Wheels rotating"

class Transmission:
    def shift_gear(self, gear):
        return f"Shifted to gear {gear}"

class Car:
    def __init__(self, engine, wheels, transmission):
        self.engine = engine  # Car has an engine
        self.wheels = wheels  # Car has wheels
        self.transmission = transmission  # Car has a transmission

    def start_car(self):
        engine_start_result = self.engine.start()
        wheels_rotation_result = self.wheels.rotate()
        return f"{engine_start_result}\n{wheels_rotation_result}"

    def drive(self, gear):
        gear_shift_result = self.transmission.shift_gear(gear)
        return f"{gear_shift_result}\n{self.wheels.rotate()}"

    def stop_car(self):
        engine_stop_result = self.engine.stop()
        return engine_stop_result

# Example Usage
car_engine = Engine()
car_wheels = Wheels()
car_transmission = Transmission()

my_car = Car(engine=car_engine, wheels=car_wheels, transmission=car_transmission)

# Start the car
print(my_car.start_car())

# Drive the car
print(my_car.drive(gear=1))
print(my_car.drive(gear=2))

# Stop the car
print(my_car.stop_car())
```

```python
##11. How can you encapsulate and hide the details of composed objects in Python classes to maintain abstraction?

'''
In Python, encapsulation is a fundamental principle that involves bundling the data (attributes) and methods that operate on the data into 
a single unit, i.e., a class. When it comes to composed objects within a class, encapsulation can be maintained by following these principles:

Private Attributes: Declare attributes as private using a single or double underscore (_ or __). This convention indicates that the attribute is 
intended to be private, and its details should not be accessed directly from outside the class.

Getters and Setters: Provide getter and setter methods to control access to the attributes. This allows you to enforce certain rules or validations
when getting or setting values.

Access Control: Utilize access modifiers (public, private, protected) to control the visibility of methods and attributes. While Python doesn't enforce 
strict access control like some other languages, naming conventions and documentation are used to signal intended access levels.

'''
##Example:

class Engine:
    def __init__(self, fuel_type):
        self.__fuel_type = fuel_type  # Private attribute

    def start(self):
        return f"Engine started. Fuel type: {self.__fuel_type}"

class Car:
    def __init__(self, make, model, engine):
        self.__make = make  # Private attribute
        self.__model = model  # Private attribute
        self.__engine = engine  # Composed object, private attribute

    def get_make(self):
        return self.__make  # Getter method

    def get_model(self):
        return self.__model  # Getter method

    def start_car(self):
        return f"{self.__make} {self.__model} - {self.__engine.start()}"

# Example Usage
car_engine = Engine(fuel_type="Gasoline")
my_car = Car(make="Toyota", model="Camry", engine=car_engine)

# Accessing attributes through getter methods
print(f"Make: {my_car.get_make()}")
print(f"Model: {my_car.get_model()}")

# Starting the car
print(my_car.start_car())
```

```python
##12. Create a Python class for a university course, using composition to represent students, instructors, and course materials.

class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name

    def enroll(self, course):
        course.add_student(self)
        return f"{self.name} enrolled in {course.title}"

class Instructor:
    def __init__(self, instructor_id, name):
        self.instructor_id = instructor_id
        self.name = name

    def assign_course(self, course):
        course.assign_instructor(self)
        return f"{self.name} assigned as the instructor for {course.title}"

class Course:
    def __init__(self, title):
        self.title = title
        self.students = []
        self.instructor = None

    def add_student(self, student):
        self.students.append(student)

    def assign_instructor(self, instructor):
        self.instructor = instructor

    def display_info(self):
        student_names = [student.name for student in self.students]
        instructor_name = self.instructor.name if self.instructor else "No instructor assigned"

        print(f"Course: {self.title}")
        print(f"Students: {', '.join(student_names)}")
        print(f"Instructor: {instructor_name}")

# Example Usage
math_course = Course(title="Mathematics")

alice = Student(student_id="S001", name="Alice")
bob = Student(student_id="S002", name="Bob")

john = Instructor(instructor_id="I001", name="John")

alice.enroll(course=math_course)
bob.enroll(course=math_course)
john.assign_course(course=math_course)

math_course.display_info()
```

```python
##14. Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes, and ingredients.

class Ingredient:
    def __init__(self, name):
        self.name = name

class Dish:
    def __init__(self, name, ingredients):
        self.name = name
        self.ingredients = ingredients

    def display_info(self):
        ingredient_names = [ingredient.name for ingredient in self.ingredients]
        print(f"Dish: {self.name}")
        print(f"Ingredients: {', '.join(ingredient_names)}")

class Menu:
    def __init__(self, name, dishes):
        self.name = name
        self.dishes = dishes

    def display_info(self):
        print(f"Menu: {self.name}")
        for dish in self.dishes:
            dish.display_info()
            print()

# Example Usage
ingredient1 = Ingredient(name="Tomato")
ingredient2 = Ingredient(name="Cheese")
ingredient3 = Ingredient(name="Bread")

dish1 = Dish(name="Margherita Pizza", ingredients=[ingredient1, ingredient2, ingredient3])
dish2 = Dish(name="Caprese Salad", ingredients=[ingredient1, ingredient2])

menu = Menu(name="Italian Menu", dishes=[dish1, dish2])

menu.display_info()
```

```python
##16. Create a Python class for a computer game character, using composition to represent attributes like weapons, armor, and inventory.

class Weapon:
    def __init__(self, name, damage):
        self.name = name
        self.damage = damage

class Armor:
    def __init__(self, name, defense):
        self.name = name
        self.defense = defense

class Inventory:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

class GameCharacter:
    def __init__(self, name, health):
        self.name = name
        self.health = health
        self.weapon = None  # Composition: GameCharacter "has-a" Weapon
        self.armor = None   # Composition: GameCharacter "has-a" Armor
        self.inventory = Inventory()  # Composition: GameCharacter "has-a" Inventory

    def equip_weapon(self, weapon):
        self.weapon = weapon

    def equip_armor(self, armor):
        self.armor = armor

    def take_damage(self, damage):
        if self.armor:
            damage -= self.armor.defense
            damage = max(0, damage)  # Ensure damage doesn't go below zero

        self.health -= damage
        print(f"{self.name} took {damage} damage. Current health: {self.health}")

    def display_inventory(self):
        print(f"{self.name}'s Inventory:")
        for item in self.inventory.items:
            print(f"- {item.name}")

# Usage
sword = Weapon("Sword", 10)
shield = Armor("Shield", 5)

player = GameCharacter("Hero", 100)
player.equip_weapon(sword)
player.equip_armor(shield)

enemy = GameCharacter("Enemy", 50)

player.take_damage(8)
player.inventory.add_item(sword)
player.inventory.add_item(shield)
player.display_inventory()
```

```python
##17. Describe the concept of "aggregation" in composition and how it differs from simple composition.

'''
Aggregation in Composition:

Aggregation is a specific form of composition where one class represents a whole (container) and another class represents a part. Unlike simple 
composition, aggregation implies a weaker relationship between the container and its parts. The key difference lies in the lifecycle and ownership
of the objects.

In aggregation:

Weak Ownership:
The container class holds a reference to the part class, but the part class can exist independently of the container.

Independent Lifecycle:
The part class can be created, modified, or destroyed without directly affecting the existence of the container.

"Has-a" Relationship:
Aggregation still represents a "has-a" relationship, but with a more relaxed connection compared to simple composition.



'''

#EXAMPLE:

class Department:
    def __init__(self, name):
        self.name = name
        self.employees = []  # Aggregation: Department "has-a" list of employees

    def add_employee(self, employee):
        self.employees.append(employee)

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

# Usage
department = Department("Engineering")

employee1 = Employee("Alice", 60000)
employee2 = Employee("Bob", 55000)

department.add_employee(employee1)
department.add_employee(employee2)

# Aggregation allows employees to exist independently of the department

```
